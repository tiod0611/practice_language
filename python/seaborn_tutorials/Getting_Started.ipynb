## Using Seaborn with Matplotlib

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# loading dataset
data = sns.load_dataset("iris")
```

```python
# draw lineplot
sns.lineplot(data=data, x="sepal_width", y="sepal_length")

# setting the title using Matplotlib
plt.title("Title using Matplotlib fucntion")
plt.show()
```

그럼 여기서 lineplot에 대해 짚고 넘어가보자.   
lineplot은 굵은 선과 그 선을 감싸는 영역으로 표시된다.  
가운데 선은 x값에 해당하는 데이터의 평균값을 의미하고 넓은 영역은 데이터의 분포다.  
이제 생각하면 쉽지?

### Example 2: Setting the xlim and ylim

### Customizing Seaborn Plots

Seaborn에서는 5가지의 테마가 가능하다
+ darkgrid
+ whitegrid
+ dark
+ white
+ ticks

```python
# set seaborn style
sns.lineplot(data=data, x="sepal_length", y="sepal_width")
sns.set_style("darkgrid")
plt.show()
```

```python
# set seaborn style
sns.lineplot(data=data, x="sepal_length", y="sepal_width")
sns.set_style("whitegrid")
plt.show()
```

```python
# set seaborn style
sns.lineplot(data=data, x="sepal_length", y="sepal_width")
sns.set_style("dark")
plt.show()
```

```python
# set seaborn style
sns.lineplot(data=data, x="sepal_length", y="sepal_width")
sns.set_style("white")
plt.show()
```

```python
# set seaborn style
sns.lineplot(data=data, x="sepal_length", y="sepal_width")
sns.set_style("ticks")
plt.show()
```

white와 ticks의 차이는 축에 눈금이 있는지 없는지다.

### Removal of spines
Spines는 데이터를 둘러싼 경계선을 의미한다. 이걸 없애는 설정도 가능하다

```python
sns.lineplot(data=data, x="sepal_length", y="sepal_width")

# removing the spines
sns.despine()
plt.show()
```

despine()을 적용함으로써 오른쪽과 위의 경계선이 사라졌다.
