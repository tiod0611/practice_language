```python
import seaborn as sns
```

```python
# Apply the default theme
sns.set_theme()
```

```python
# Load an example dataset
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
# Create a visualization
# 시각화를 할 때 떠올려야 할 항목
# 1. 무슨 데이터를 사용할거지?
# 2. x 축은?
# 3. y 축은?
# 4. 특정 카테고리를 구별해서 보고싶다면?
# 5. 여러 칼럼에 대해서 구별하면?
# 이제는 데이터 스타일에 대해서 생각해보자.
# 
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", # 데이터의 구분. 서로 다른 색으로 구분됨.
    style="smoker",
    size="size",
)
```
