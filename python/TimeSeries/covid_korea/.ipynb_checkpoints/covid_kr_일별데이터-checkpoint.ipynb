```python
# 분리된 날짜 데이터 통합
# 데이터 특징을 보니, 같은 날 다른 시간대 결과도 누적이므로 같은 날 데이터 중 더 큰 값을 남기자.

# 중복 확인
covid_kr[covid_kr.duplicated(['stateDt', 'stateTime']) == True]
```

```python
# 중복 데이터 제거
covid_kr.drop_duplicates(['stateDt', 'stateTime'], inplace=True)
covid_kr = covid_kr.reset_index(drop=True)
```

```python
# 분리된 날자 통합하기

covid_kr[covid_kr.duplicated(['stateDt']) == True]
```

```python
covid_day = covid_kr.groupby('stateDt').max()
covid_day.reset_index(inplace=True) # 날짜가 index로 가면서 자연스럽게 역순 배열이 됨.
covid_day.head()
```

```python
# 중복 확인
covid_day[covid_day.duplicated(['stateDt'])==True]
```

```python
# 날자 통합이 되었으므로 stateTime은 제거한다.
covid_day.drop(columns=['stateTime'], inplace=True)
```

```python
covid_day.head()
```

## 결측치 처리

이제 결측치를 처리해보자.  
decideCnt, examCnt, deathCnt 모두 누적 데이터이므로 NaN이라면 이전 행의 값으로 치환하는 것으로 처리하자.  
앞 행의 값으로 대체하는 작업은 '보간법'으로 할 수 있다.  
단, examCnt의 경우 2021년 12월 04일 이후로는 의도적으로 기입하지 않았기 때문에 보류한다.

```python
covid_day[covid_day['examCnt'].isna() == True].head(10)
```

```python
# 기입 오류 처리
covid_day.loc[covid_day['deathCnt'] == 1, 'deathCnt'] = 2
```

```python
covid_day[covid_day['deathCnt'] == 1]
```

```python
covid_day['decideCnt'] = covid_day['decideCnt'].interpolate(method='linear')
covid_day['deathCnt'] = covid_day['deathCnt'].interpolate(method='linear')
covid_day_intpol = covid_day.round(0)
covid_day_intpol.head(20)
```

```python
# 1행의 데이터는 0으로 하자.
covid_day_intpol.iloc[0] = covid_day_intpol.iloc[0].fillna(0)

# index가 656 이후로는 examCnt가 없다. 이 값은 건들이지 않도록 하자.
covid_day_intpol.iloc[0:5] = covid_day_intpol.iloc[0:5].interpolate(method='linear').round(0)
covid_day_intpol.head(20)
```

# 신규 확진자, 신규 사망자, 사망율, 확진률 데이터 구하기

이제 decideCnt와 deathCnt를 차분해서 일별 신규 데이터를 구해보자.

```python
covid_day_intpol['newDecideCnt'] = covid_day_intpol['decideCnt'].diff(1)
covid_day_intpol['newDeathCnt'] = covid_day_intpol['deathCnt'].diff(1)
```

```python
covid_day_intpol.head(20)
```

 검사자 중 코로나 확진자 비율도 구해보자


```python
covid_day_intpol['decideRate'] = round(covid_day_intpol['newDecideCnt'].div(covid_day_intpol['examCnt']) * 100, 2)
covid_day_intpol['deathRate'] = round(covid_day_intpol['newDeathCnt'].div(covid_day_intpol['decideCnt']) * 100, 2)
```

```python
# 첫행 처리
covid_day_intpol.iloc[0] = covid_day_intpol.iloc[0].fillna(0)
```

# 데이터 시각화

```python
# 백신 접종 시작일 index 구하기
print(covid_day_intpol[covid_day_intpol['stateDt'] == 20210227].index[0])
print(covid_day_intpol[covid_day_intpol['stateDt'] == 20210101].index[0])
print(covid_day_intpol[covid_day_intpol['stateDt'] == 20220101].index[0])
print(covid_day_intpol[covid_day_intpol['stateDt'] == 20230101].index[0])
```

```python
titles = ['decideCnt', 'examCnt', 'deathCnt', 'newDecideCnt', 'newDeathCnt', 'decideRate', 'deathRate'][::-1]
vaccine = covid_day_intpol[covid_day_intpol['stateDt'] == 20210227].index[0]
newYears = [335, 680, 1018]
lenTitle = len(titles)
fig, axs = plt.subplots(lenTitle, 1, figsize=(20, 40))

# fig.suptitle("연령대 별 사망율 추이")

for i in range(lenTitle):
    column = titles.pop()
    axs[i].plot(covid_day_intpol[column], linewidth='3')
    axs[i].set_title(column, fontsize=20)
    axs[i].axvline(x=vaccine, color='r', linestyle='--', linewidth=2) # 백신 배포일
    for j in range(3):
        if column not in ['examCnt', 'decideRate']:
            axs[i].axvline(x=newYears[j], color='orange')
plt.show()
```

확진율과 사망율에서 너무 높은 수치가 있어서 그래프가 왜곡되어 보인다. 이를 처리해주자.

그리고 신규 사망자 데이터를 보니, 신년을 기준으로 사망자 비율이 비교적 높아보인다.

```python
fig, axs = plt.subplots(6, 1, figsize=(10, 20)) 
flag = [121, 335, 483, 680, 817, 1018]
season = ['2020_summer', '2020_winter', '2021_summer', '2021_winter', '2022_summer', '2022_winter']
interval = 60

for i in range(len(flag)):
    
    if (flag[i]-interval) <= 0:
        idx_s = 0
    else:
        idx_s = flag[i] - interval
    if (flag[i]+interval) >= len(covid_day_intpol['newDeathCnt']):
        idx_e = len(covid_day_intpol['newDeathCnt'])
    else:
        idx_e = flag[i] + interval
        
    axs[i].plot(covid_day_intpol.iloc[idx_s:idx_e]['newDeathCnt'], linewidth='2')
    axs[i].set_title(season[i], fontsize=10)
    axs[i].axvline(x=flag[i], color='orange')
    axs[i].annotate(
        str(covid_day_intpol.iloc[flag[i]]['newDeathCnt'])[:-2],
        ha='center', va='bottom',
        xytext=(flag[i] + (idx_e - flag[i])*0.15 , 
                covid_day_intpol.iloc[flag[i]]['newDeathCnt']*1.1),
        fontsize=20,
        color='r',
        backgroundcolor='white',
        xy = (flag[i], 
              covid_day_intpol.iloc[flag[i]]['newDeathCnt'] ),
        arrowprops={'facecolor': 'r',}
    )
    
```

데이터를 보아 겨울철에 사망자 수가 상대적으로 올라가는 것으로 볼 수 있다.
