# Activate Functions!

```python
import torch
import torch.nn as nn

import matplotlib.pyplot as plt
```

```python
x = torch.sort(torch.randn(100) * 10)[0] # 정규 분포 샘플값을 만든다음 10을 곱해서 가져온다. [0]은 sorted value [1]은 indice
x
```

### Sigmoid
$$ \sigma(x)=\frac{1}{1+e^{-x}} $$

```python
# 객채로 사용하기
act = nn.Sigmoid()
print(act(x))
```

```python
# 함수로 사용하기
print(torch.sigmoid(x))
```

```python
# 시각회
plt.plot(x.numpy(), torch.sigmoid(x).numpy())
plt.show()
```

### Hyperbolic Tangent(Tanh)

$ tanh(x) = \frac{e^{x}-e^{-x}}{e^{x}+e^{-x}} $

```python
# 객체로 만들어서 사용
act = nn.Tanh()
print(act(x))
```

```python
print(torch.tanh(x))
```

```python
plt.plot(x, torch.tanh(x))
plt.show()
```
