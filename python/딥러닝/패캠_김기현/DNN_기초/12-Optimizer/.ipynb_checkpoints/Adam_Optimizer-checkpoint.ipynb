# Adam Optimizer

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.datasets import fetch_california_housing
california = fetch_california_housing()
```

```python
df = pd.DataFrame(california.data, columns=california.feature_names)
df['Target'] = california.target
df.tail()
```

```python
scaler = StandardScaler()
scaler.fit(df.values[:, :-1])
std_tr = scaler.transform(df.values[:, :-1])
df = pd.DataFrame(std_tr, columns=california.feature_names)
df['Target'] = california.target
df.head()
```

```python
df.tail()
```

## Train Model with PyTorch

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
data = torch.from_numpy(df.values).float()
data.shape
```

```python
x = data[:, :-1]
y = data[:, -1:]
```

```python
# figures
n_epochs = 4000
batch_size = 256
print_interval = 100
# learning_rate = 1e-2 # 하지만 사용하지 않음
```

## Build Model

```python
model = nn.Sequential(
    nn.Linear(x.size(-1), 6),
    nn.LeakyReLU(),
    nn.Linear(6, 5),
    nn.LeakyReLU(),
    nn.Linear(5, 4),
    nn.LeakyReLU(),
    nn.Linear(4, 3),
    nn.LeakyReLU(),
    nn.Linear(3, y.size(-1)),
)
```

```python
model
```

```python
# We don't need learning_rate hyper-parameter.
optimizer = optim.Adam(model.parameters())
```

```python
for i in range(n_epochs):
    indices = torch.randperm(x.size(0))
    x_ = torch.index_select(x, dim=0, index=indices)
    y_ = torch.index_select(y, dim=0, index=indices)
    
    x_ = x_.split(batch_size, dim=0)
    y_ = y_.split(batch_size, dim=0)
    
    y_hat = []
    total_loss = 0
    
    for x_i, y_i in zip(x_, y_):
        y_hat_i = model(x_i)
        loss = F.mse_loss(y_hat_i, y_i)
        
        optimizer.zero_grad()
        loss.backward()
        
        optimizer.step()
        
        total_loss += float(loss) # This is very important to prevent memory leak.
        y_hat += [y_hat_i]
        
    total_loss = total_loss / len(x_)
    if (i+1) % print_interval == 0:
        print(f"Epochs {i+1}: loss={total_loss:.5f}")
```

```python
y_hat = torch.cat(y_hat, dim=0)
y = torch.cat(y_, dim=0)
```

# Let's see the result!

```python
df = pd.DataFrame(torch.cat([y, y_hat], dim=1).detach().numpy(), columns=['y', 'y_hat'])

sns.pairplot(df, height=5)
plt.show()
```
