## The object-oriented interface and the pyplot interface

```python
x = np.linspace(0, 2, 100)

# Note that even in the 00-style, we use '.pyplot.figure' to create the figure
fig, ax = plt.subplots() # Plot some data on the axes. # figure 
ax.plot(x, x, label='linear') # Plot some data on the axes.
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2, label='quadratic') # Plot more data on the axes
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**3, label='cubic')
```

```python
fig, ax = plt.subplots()
ax.plot(x, x, label='linear') # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic') # Plot some data on the axes.
ax.plot(x, x**3, label='cubic') # Plot some data on the axes.
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Simple Plot')
ax.legend()
```

or pyplot style

```python
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Simple Plot')
plt.legend()
plt.show()
```

```python
def my_plotter(ax, data1, data2, param_dict):
    """
    Parameters
    ------------
    ax: Axes
        The axes to draw to
        
    data1: array
        The x data
        
    data2: array
        The y data
        
    pram_dict: dict
        Dictionary of kwargs to pass to ax.plot
        
    ========
    returns
    out: list
        list of artists added
    """
    
    out = ax.plot(data1, data2, **param_dict)
    return out
```

```python
data1, data2, data3, data4 = np.random.rand(4, 100)
data1.shape, data2.shape, data3.shape, data4.shape
```

```python
fig, ax = plt.subplots()
my_plotter(ax, data1, data2, {'marker': 'x'})
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})
```
