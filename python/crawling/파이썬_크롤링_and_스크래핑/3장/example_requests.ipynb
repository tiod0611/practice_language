```python
import requests

r = requests.get('http://hanbit.co.kr') # get() 함수로 웹 페이지를 추출함

```

```python
print(type(r))

```

```python
print(r.status_code) # status_code 속성으로 HTTP 상태 코드를 확인함

```

```python
print(r.headers['content-type']) # headers 속성으로 HTTP 헤더 딕셔너리로 추출

```

```python
print(r.encoding) # encoding 속성으로 HTTP헤더를 기반으로 인코딩 추출함

```

```python
print(r.content) # content 속성으로 bytes 자료형의 응답 본문을 추출함

```

```python
print(r.text) # text 속성으로 str 자료형으로 디코딩된 응답 본문을 추출함
```
