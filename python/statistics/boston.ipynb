다음의 가설에 대해, 상관분석을 수행하고 상관계수를 구하시오.  (소수점 4자리에서 반올림하시오)  
   tax(세율) --> medv(집값)"


```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

import scipy.stats as stats
```

```python
boston = pd.read_csv("https://bit.ly/3PArw4W")
```

```python
boston.head()
```

```python
sns.regplot(data=boston, x=boston['tax'], y=boston['medv'])
plt.show()
```

```python
sns.jointplot(data=boston, x=boston['tax'], y=boston['medv'])
plt.show()
```

```python
stats.pearsonr(boston['tax'], boston['medv'])
```
