```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style(style='whitegrid')

#폰트 설정
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False # 한글폰트로 -기호를 사용할 때, ascii 문자로 표시함.
```

```python
# 정규 분포를 따르는 데이터에서 추출한 표본의 크기에 따라 표본 평균이 변하는 모습을 시각화

sample_means = []
for i in range(2, 10001):
    samples = np.random.normal(170, 7.0, i)
    sample_mean = np.mean(samples)
    sample_means.append(sample_mean)
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=range(2, 10001), y=sample_means)
plt.axhline(y=170, color='red', linestyle='--', label='모평균 170')
plt.title("(큰수 법칙)-표본 집단 크기에 따른 표본 평균의 변화")
plt.legend()
plt.show()
```

```python
# 하지만 n이 무한이 커지더라도 모평균과 동일해진다는 법은 없다. 다만 근사할 뿐이다.
```

```python
# 중심극한정리
# 위 데이터를 histogram으로 시각화하여 표본 평균들의 분포를 살펴보자.
# 이 분포가 정규분포 모양을 하고 있을까?
```

```python
sns.histplot(x=sample_means, color='skyblue')
plt.xlim(160, 180)
plt.title("표본 평균")
plt.show()
```
