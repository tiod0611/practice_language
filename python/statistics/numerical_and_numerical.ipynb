```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

sns.set_style(style='whitegrid')
```

```python
# set data
temperature = [25, 26, 28, 21, 12, 14, 14, 31, 1, 32]
sales = [535, 625, 661, 488, 316, 368, 309, 697, 55, 738]

df = pd.DataFrame(data={'temperature':temperature, 'sales':sales},)
df
```

```python
# 시각적으로 살펴보기 
# 1. 산점도
# 2. 상관계수의 직선

sns.scatterplot(data=df, x='temperature', y='sales')
plt.show()
```

```python
# scatter와 함께 회귀선 그리기.
sns.regplot(data=df, x='temperature', y='sales')
plt.show()
```

```python
df.corr()
```

```python
sns.pairplot(df, kind="reg", palette="Set2")
plt.show()
```
