```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4]) # plot에 하나의 시퀀스만 넣으면 이건 y값으로 본다. 그리고 x값은 0부터 시작함
plt.ylabel('just series number')
plt.show()
```

```python
plt.plot([1,2,3,4], [3,4,5,6]) #두개의 시퀀스를 넣을 경우 앞에서부터 x, y다. 
plt.show()
```

# plotting with keyword strings

```python
df = {'a': np.random.randn(50),
       'c': np.random.randint(0, 60, 50), #(xmin, xmax, number of)
       'd': np.random.randn(50)}
df
```

```python
df['b'] = df['a'] + 10 * np.random.randn(50)
df['d'] = np.abs(df['d']) *100
df
```

```python
plt.scatter('a', 'b', c='c', s='d', data=df) # c=color, s=scale, data=사용할 데이터 변수 이름
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```
