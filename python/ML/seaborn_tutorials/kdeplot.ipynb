```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('penguins')
data.head()
```

```python
sns.kdeplot(data=data, x='body_mass_g', color='black')
sns.kdeplot(data=data, x='body_mass_g', hue='species', fill=True)
plt.xlim(1000, 10000)
plt.ylim(0, 0.001)
plt.show()
```

```python
sns.kdeplot(data=data, x='body_mass_g', hue='sex')
plt.xlim(1000, 10000)
plt.ylim(0, 0.001)
plt.show()
```
