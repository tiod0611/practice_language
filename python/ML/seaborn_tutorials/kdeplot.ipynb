```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("iris")
```

```python
# 기본 KDE
# 단일 변수의 분포를 부드러운 곡선으로 시각화 하고 싶을 때 사용함
# 곡선의 높이는 해당 지점의 데이터 밀도를 나타냄. 
# 곡선 아래 면적은 전체 데이터셋에서 그 구간의 데이터가 차지하는 비율을 의미
# 곡선이 높고 좁으면 데이터가 그 지점에 집중되어 있음을 의미함
# 곡선이 낮고 넓으면 데이터가 넓게 분산되어 있음을 나타냄.
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
sns.kdeplot(data['sepal_length'])
plt.grid()

plt.subplot(1, 3, 2)
sns.kdeplot(data['sepal_width'])
plt.grid()

plt.subplot(1, 3, 3)
sns.kdeplot(data=data, x='sepal_width', bw_adjust=0.5) # 세부적인 데이터 특징을 포착하기 위함
plt.grid()


plt.tight_layout()
plt.show()
```

```python
# 데이터의 분포를 연속적인 선으로 보여줌.


# 다른 plot과 비교해볼까
```

```python
sns.boxplot(data['sepal_length'])
plt.show()
```

```python
sns.histplot(data['sepal_length'])
plt.grid()
plt.show()
```

```python
sns.lineplot(data['sepal_length'])
plt.grid()
plt.show()
```

```python
# 두 변수 간에 관계를 볼 수 있는 2D KDE plot
# 두 데이터의 밀도를 표현함.
# 등고선이 밀집한 영역은 해당 위치에 데이터가 몰려있다는 것을 의미함.

sns.kdeplot(data=data, x='sepal_width', y='sepal_length')
plt.grid()
plt.xlim(1, 5)
plt.ylim(3, 9)
plt.show()
```

```python
sns.scatterplot(data=data, x='sepal_width', y=data['sepal_length'])
plt.grid()
plt.xlim(1, 5)
plt.ylim(3, 9)
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.kdeplot(data=data, x='sepal_width', y='sepal_length')
plt.grid()
plt.xlim(1, 5)
plt.ylim(3, 9)

plt.subplot(1, 2, 2)
sns.scatterplot(data=data, x='sepal_width', y=data['sepal_length'])
plt.grid()
plt.xlim(1, 5)
plt.ylim(3, 9)
plt.show()
```

```python
# KDE 플롯에 색상 추가하기
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
sns.kdeplot(data=data, x='sepal_width', hue='species')
plt.grid()

plt.subplot(1, 3, 2)
sns.kdeplot(data['sepal_width'])

plt.subplot(1, 3, 3)
sns.histplot(data=data, x=data['sepal_width'], hue='species')
# hue를 적용할 때, data와 x 변수를 모두 제공해야 오류가 나지 않는다. 
plt.grid()

plt.tight_layout()
plt.show()
```
