```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('penguins')
data.head()
```

```python
sns.kdeplot(data=data, x='body_mass_g', color='black')
sns.kdeplot(data=data, x='body_mass_g', hue='species', fill=True)
plt.xlim(1000, 10000)
plt.ylim(0, 0.001)
plt.show()
```

```python
sns.kdeplot(data=data, x='body_mass_g', hue='sex')
plt.xlim(1000, 10000)
plt.ylim(0, 0.001)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("iris")
```

```python
#  KDE
#          
#       . 
#           
#         
#        .
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
sns.kdeplot(data['sepal_length'])
plt.grid()

plt.subplot(1, 3, 2)
sns.kdeplot(data['sepal_width'])
plt.grid()

plt.subplot(1, 3, 3)
sns.kdeplot(data=data, x='sepal_width', bw_adjust=0.5) #     
plt.grid()


plt.tight_layout()
plt.show()
```

```python
#     .


#  plot 
```

```python
sns.boxplot(data['sepal_length'])
plt.show()
```

```python
sns.histplot(data['sepal_length'])
plt.grid()
plt.show()
```

```python
sns.lineplot(data['sepal_length'])
plt.grid()
plt.show()
```

```python
#        2D KDE plot
#    .
#         .

sns.kdeplot(data=data, x='sepal_width', y='sepal_length')
plt.grid()
plt.xlim(1, 5)
plt.ylim(3, 9)
plt.show()
```

```python
sns.scatterplot(data=data, x='sepal_width', y=data['sepal_length'])
plt.grid()
plt.xlim(1, 5)
plt.ylim(3, 9)
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.kdeplot(data=data, x='sepal_width', y='sepal_length')
plt.grid()
plt.xlim(1, 5)
plt.ylim(3, 9)

plt.subplot(1, 2, 2)
sns.scatterplot(data=data, x='sepal_width', y=data['sepal_length'])
plt.grid()
plt.xlim(1, 5)
plt.ylim(3, 9)
plt.show()
```

```python
# KDE   
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
sns.kdeplot(data=data, x='sepal_width', hue='species')
plt.grid()

plt.subplot(1, 3, 2)
sns.kdeplot(data['sepal_width'])

plt.subplot(1, 3, 3)
sns.histplot(data=data, x=data['sepal_width'], hue='species')
# hue  , data x      . 
plt.grid()

plt.tight_layout()
plt.show()
```
