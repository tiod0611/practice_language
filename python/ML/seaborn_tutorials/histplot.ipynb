## Histograms !
단변량 변수의 분포를 시각적으로 파악도록 도와주는 그래프!

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme()
```

```python
olympic_df = pd.read_csv('./data/olympic_athletes/athlete_events.csv')
web_log_df = pd.read_csv('./data/web_log/train.csv')
```

```python
display(olympic_df)
display(web_log_df)
```

### 기본적인 histgram 사용법 `sns.histplot(data, x or y, discrete, bins)`

```python
# Example 1
sns.histplot(data=olympic_df, x='Height')
```

```python
sns.histplot(data=web_log_df, x = 'TARGET')
```

```python
# Example 2
sns.histplot(data=olympic_df, y='Height')
plt.show()
```

```python
# Example 3 : Custom bins
sns.histplot(data=olympic_df,x = 'Height', bins=10)
```

```python
# Example 3.1 : bins range
sns.histplot(data=olympic_df, x='Height', bins=[100, 120, 140, 160, 180, 200, 250])
```

```python
# Example 4 : descrete=True
# 히스토그램은 기본적으로 값의 범위에 막대가 그려지게 된다. 위 예시를 보면 140-160 사이에 막대가 있다는 것을 알 수 있다.
# discrete=True는 빈도수를 정확하게 알 수 있는 매개변수다. 이를 적용해보자.
sns.histplot(data=olympic_df, x='Height', discrete=True)
plt.show()

# 단, discrete=True를 사용하면 bins 매개변수를 사용할 수 없다. 
# discrete는 데이터 분포를 정확하게 알 고 싶을 때 사용할 수 있겠다.
```

```python
# Example 5: kernel density estimate(`kde`)
# 데이터셋의 확률밀도함수를 '추정'하는 Kernel Density Estimate를 플로팅
sns.histplot(data=olympic_df, x='Height', 
             discrete=True, 
             kde=True)
plt.show()
```

```python
sns.histplot(data=olympic_df, x='Height', 
             kde=True)
plt.show()
```

```python
# Example 6 : hue
# 데이터에 카테고리로 구분한다.
sns.histplot(data=olympic_df, x='Height', hue='Sex')
```

```python
# multiple은 hue를 사용했을 때, 데이터를 어떻게 표현할지를 결정한다.
# 기본은 'layer'로 데이터를 겹쳐서 보여준다.
# 'dodge'로 하면 겹치지 않고 나란히 보여준다.
sns.histplot(data=olympic_df, x='Height', bins=10, hue='Sex', multiple='layer')
plt.show()
```

```python
sns.histplot(data=olympic_df, x='Height', bins=10, hue='Sex', multiple='dodge')
plt.show()
```

```python
# Example 8 : y값을 Count가 아니라 비율로 보는 법 stat='probability'
sns.histplot(data=olympic_df, x='Height', bins=10, hue="Sex", multiple='dodge', stat="probability")
```

```python
# Bounus. Histogram을 사용해서 heatmap을 그려보자.
olympic_df['Weight'] = olympic_df['Weight'].astype(float)
olympic_df['Height'] = olympic_df['Height'].astype(float)

sns.histplot(data=olympic_df, x="Height", y='Weight')
plt.show()
```

```python
sns.histplot(data=olympic_df, x="Sex", y='Height', hue="Sex", legend=False)
```
