## Using Seaborn with Matplotlib

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# loading dataset
data = sns.load_dataset("iris")
```

```python
# draw lineplot
sns.lineplot(data=data, x="sepal_width", y="sepal_length")

# setting the title using Matplotlib
plt.title("Title using Matplotlib fucntion")
plt.show()
```

### Customizing Seaborn Plots

Seaborn에서는 5가지의 테마가 가능하다
+ darkgrid
+ whitegrid
+ dark
+ white
+ ticks

```python
# set seaborn style
sns.lineplot(data=data, x="sepal_length", y="sepal_width")
sns.set_style("darkgrid")
plt.show()
```

```python
# set seaborn style
sns.lineplot(data=data, x="sepal_length", y="sepal_width")
sns.set_style("whitegrid")
plt.show()
```

```python
# set seaborn style
sns.lineplot(data=data, x="sepal_length", y="sepal_width")
sns.set_style("dark")
plt.show()
```

```python
# set seaborn style
sns.lineplot(data=data, x="sepal_length", y="sepal_width")
sns.set_style("white")
plt.show()
```

```python
# set seaborn style
sns.lineplot(data=data, x="sepal_length", y="sepal_width")
sns.set_style("ticks")
plt.show()
```

white와 ticks의 차이는 축에 눈금이 있는지 없는지다.

### Removal of spines
Spines는 데이터를 둘러싼 경계선을 의미한다. 이걸 없애는 설정도 가능하다

```python
sns.lineplot(data=data, x="sepal_length", y="sepal_width")

# removing the spines
sns.despine()
plt.show()
```

despine()을 적용함으로써 오른쪽과 위의 경계선이 사라졌다.

### Changing the figure size

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('iris')
```

```python
# changing the figure size
plt.figure(figsize = (2, 4)) # 단위는 inch. 
                            #하지만 단위를 생각하기 보다 단순히 그래프의 비율로 생각하는 편이 낫다
sns.lineplot(data=data, x='sepal_length', y='sepal_width')
# remove the spines
sns.despine()
plt.show()
```

```python
# figure 사이즈를 고정하고 싶다면 rcParams를 사용하면 된다. 그럼 이후 그래프는
# 해당 사이즈로 고정되어 나온다.
# rc는 Running time Configuration으로 현재 런타임에서 작동하는 matplotlib스타일을 설정하는 방법이다.
# 이런 설정은 running time이 살아있는 동안 유지된다.
plt.rcParams['figure.figsize'] = [2, 4] 
sns.lineplot(data=data, x='sepal_length', y='sepal_width')
plt.show()
```

### Plot의 기본적인 scale을 설정
set_context()를 사용하면 전체적인 scale을 설정하기 쉽다.  
context 값으로는 'notebook', 'paper', 'talk', 'poster'가 있는데, 각각 해당 플랫폼에 적절한 크기다.

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset('iris')
```

```python
# scale 적용해보기
sns.lineplot(data=data, x='sepal_length', y='sepal_width')
sns.set_context('paper') # 논문에 적합한 scale
plt.show()
```

```python
# scale 적용해보기
sns.lineplot(data=data, x='sepal_length', y='sepal_width')
sns.set_context('notebook') # 논문에 적합한 scale
plt.show()
```

```python
# scale 적용해보기
sns.lineplot(data=data, x='sepal_length', y='sepal_width')
sns.set_context('poster') # 논문에 적합한 scale
plt.show()
```

```python
# scale 적용해보기
sns.lineplot(data=data, x='sepal_length', y='sepal_width')
sns.set_context('talk') # 논문에 적합한 scale
plt.show()
```

## Setting the style temporarily
axes_style()은 일시적으로 스타일을 설정하는 방법이다. 

```python
sns.set_context('notebook')
```

```python
def plot():
    sns.lineplot(data=data, x='sepal_length', y='sepal_width')
    

with sns.axes_style('darkgrid'):
    # Adding the subplot
    plt.subplot(211)
    plot()

plt.subplot(212)
plot()
```

## Color Platte
Colormaps는 plot를 효과적이면서 쉽게 시각화하는 데 사용된다. 다른 종류의 plot 마다 다른 colormaps을 사용할 수 있다.  __color_palatte()__ method는 plot에 colors를 입힌다. 다른 function인 __palplot()__ 은 color palette를 다루고 color palette를 수평적 배열로 다룹니다. (?) 

```python
# 문장이 이해가 안될 때는 예제로 살펴보자.

# current color palette
palette = sns.color_palette()

# plots the color palette as a horizontal array
sns.palplot(palette)
plt.show()
```
