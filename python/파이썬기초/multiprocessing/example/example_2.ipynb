## 단순히 총 20만을 카운트 하는 예제.
이를 single process와 multi process로 나누어 시간을 측정해보자.

이 예제 코드의 출처는 [이곳](https://niceman.tistory.com/145#%EB%B3%B4%ED%86%B5%20Multiprocessing%20%EC%8B%A4%ED%96%89%20%ED%9D%90%EB%A6%84%EC%9D%80%20%EC%95%84%EB%9E%98%EC%99%80%20%EA%B0%99%EC%9D%B4%20%EB%82%98%ED%83%80%EB%82%BC%20%EC%88%98%20%EC%9E%88%EC%9C%BC%EB%A9%B0%2C%20%EA%B4%80%EB%A0%A8%20%EC%98%88%EC%A0%9C%EB%8A%94%20%ED%95%98%EB%8B%A8%EC%97%90%EC%84%9C%20%EC%84%A4%EB%AA%85%ED%95%A9%EB%8B%88%EB%8B%A4.MultiProcessing%20%EC%B2%98%EB%A6%AC%20%EA%B5%AC%EC%A1%B0%C2%A0-%20%EC%B6%9C%EC%B2%98%ED%8C%8C%EC%9D%B4%EC%8D%AC(Python)%C2%A0MultiProcessing(Pool)%C2%A0-%20%EC%98%88%EC%A0%9C%20%EC%BD%94%EB%93%9C-1)이다.

```python
import time
```

```python
# single

start_time = time.time()

# 멀티쓰레드를 사용하지 않은 경우 (20만 카운트)
def count(name):
    for i in range(1, 50001):
        if i % 100 == 0:
            print(name, " : ", i)
num_list = ['p1', 'p1', 'p3', 'p4']
for num in num_list:
    count(num)

print("----- %s seconds -----" % (time.time() - start_time))
```

```python
# multi processing
import multiprocessing

start_time =  time.time()

# 멀티쓰레드 사용하는 경우 (20만 카운트)
# Pool 사용해서 함수 실행 을 병렬
def count(name):
    for i in range(1, 50001):
        if i % 100 == 0:
            print(name, " : ", i)

num_list = ['p1', 'p2', 'p3', 'p4']

if __name__ == '__main__':
    # 멀티 쓰레딩 Pool 사용
    pool = multiprocessing.Pool(processes=2) # 현재 시스템에서 사용할 프로세스 개수
    pool.map(count, num_list)
    pool.close()
    pool.join()

print("----- %s seconds -----" %(time.time() - start_time))
```
